```python
delivery["Innings No"].unique()
delivery["Match id"].unique().size

total_score_df =delivery.groupby(["Match id","Innings No"])
total_score_df = total_score_df['runs_scored'].sum().reset_index()
total_score_df = total_score_df[total_score_df['Innings No']==1]

match_df =match.merge(total_score_df[['Match id','runs_scored']],left_on='Match_ID',right_on='Match id')

match_df
```

```python
match_req_df = match_df[['Match_ID','Venue','Match_Winner', 'Toss_Winner','runs_scored']]
```

```python
delivery_df =match_req_df.merge(delivery,left_on='Match_ID',right_on='Match id')
delivery_df =delivery_df[delivery_df['Innings No']==2]
delivery_df.columns

```

```python
delivery_df['current_score']=delivery_df.groupby('Match id')['runs_scored_y'].cumsum()
delivery_df['runs_left'] = delivery_df['runs_scored_x']-delivery_df['current_score']+1
# split the column Ball No in delivery_df with decimal point and convert both left and right side to integer
delivery_df['over'] = delivery_df['Ball No'].apply(lambda x: int(x)+1)
delivery_df['ball'] = delivery_df['Ball No'].apply(lambda x: int((x-int(x))*10)+1)

delivery_df['balls_left']=126-(delivery_df['over']*6+delivery_df['ball'])
```

```python
delivery_df['Player Out']=delivery_df['Player Out'].fillna(0)
delivery_df['player_dismissed']=delivery_df['Player Out'].apply(lambda x : x if x ==0  else 1)
wickets = delivery_df.groupby('Match id')['player_dismissed'].cumsum().values
delivery_df['wickets']=10- wickets
```

```python
delivery_df['crr']=delivery_df["current_score"]*6/(120-delivery_df["balls_left"])
delivery_df['rrr']=delivery_df["runs_left"]*6/delivery_df["balls_left"]
delivery_df = delivery_df.reset_index()
```

```python
def is_win(df):
    winner = []
    for item,row in df.iterrows():
        if row["Match_Winner"] == row["Batting team"]:
            winner.append(1)
        else:
            winner.append(0)

    return winner
            
delivery_df['winner']=is_win(delivery_df)
```

```python
final_df =delivery_df[['Match id','Batting team','Bowling team','Venue','runs_left','balls_left','wickets','runs_scored_x','crr','rrr','winner']]
final_df =final_df.sample(final_df.shape[0])
```

```python
final_df.dropna(inplace=True)
final_df =final_df[final_df.balls_left !=0]
```

```python
from sklearn.model_selection import train_test_split
X = final_df.drop(columns=['winner','Match id'])
y = final_df['winner']
X_copy =final_df.drop(columns=['winner'])
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
ohe = OneHotEncoder()
ohe.fit_transform(X.iloc[:,:3])
category = ohe.categories_
trf = ColumnTransformer([
    ('trf',OneHotEncoder(categories=category , drop='first'),['Batting team','Bowling team','Venue'])
],remainder='passthrough')
```

```python
import lightgbm as lgb
# Create an instance of the LGBMClassifier model
model = lgb.LGBMClassifier(
    boosting_type='gbdt',
    num_leaves=20,        
    max_depth=-1,        
    learning_rate=0.1,
    force_col_wise=True,
    n_estimators=100            
)



pipe = Pipeline([
    ('step1',trf),
    ('step2',model),
])
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict_proba(X_test)
# accuracy_score(y_test,y_pred)
y_pred
```
